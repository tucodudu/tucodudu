#15. Refazer o programa 14 realizando as quatro operações aritméticas básicas

```
# Solicitar os dois números reais ao usuário
numero1 = float(input("Digite o primeiro número real: "))
numero2 = float(input("Digite o segundo número real: "))

# Realizar as quatro operações aritméticas
soma = numero1 + numero2
subtracao = numero1 - numero2
multiplicacao = numero1 * numero2

# Certificar-se de que o segundo número não seja zero para evitar divisão por zero
if numero2 != 0:
    divisao = numero1 / numero2
else:
    divisao = "Não é possível dividir por zero"

# Apresentar os resultados na tela
print(f"A soma dos números {numero1} e {numero2} é igual a {soma}")
print(f"A subtração dos números {numero1} e {numero2} é igual a {subtracao}")
print(f"A multiplicação dos números {numero1} e {numero2} é igual a {multiplicacao}")
print(f"A divisão dos números {numero1} por {numero2} é igual a {divisao}")
```

#16. Escreva um programa que solicite o valor fixo do salário de um vendedor, o total vendido no mês e o percentual de comissão do vendedor. Ao final apresentar o salário bruto

```
# Solicitar o salário fixo ao vendedor
salario_fixo = float(input("Digite o salário fixo do vendedor: "))

# Solicitar o total vendido no mês
total_vendido = float(input("Digite o total vendido no mês: "))

# Solicitar o percentual de comissão do vendedor
percentual_comissao = float(input("Digite o percentual de comissão do vendedor (em porcentagem): "))

# Calcular o salário bruto
salario_bruto = salario_fixo + (total_vendido * (percentual_comissao / 100))

# Apresentar o salário bruto
print(f"O salário bruto do vendedor é: R${salario_bruto:.2f}")
```
