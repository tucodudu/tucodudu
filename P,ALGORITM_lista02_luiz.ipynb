#10. O sistema de avaliação de determinada disciplina é composto por três provas.
# A primeira prova tem peso 2,
# A segunda tem peso 3 e a terceira tem peso
#5. Considerando que a média para aprovação é 6.0, faça um algoritmo para calcular a média final de um aluno desta disciplina e dizer se o aluno foi aprovado ou não

```python
# Solicitar as notas das três provas ao aluno
nota_prova1 = float(input("Digite a nota da primeira prova (peso 2): "))
nota_prova2 = float(input("Digite a nota da segunda prova (peso 3): "))
nota_prova3 = float(input("Digite a nota da terceira prova (peso 5): "))

# Calcular a média ponderada
media_final = (nota_prova1 * 2 + nota_prova2 * 3 + nota_prova3 * 5) / 10

# Verificar se o aluno foi aprovado ou reprovado
if media_final >= 6.0:
    print(f"Média final: {media_final:.2f}. Aluno aprovado!")
else:
    print(f"Média final: {media_final:.2f}. Aluno reprovado.")
```

#14. Escreva um programa em linguagem C que solicite ao usuário a média para aprovação em um curso e em seguida solicite ao usuário o nome, sexo e as 03 notas do aluno
#e ao final imprima a frase: "O aluno XXXXX foi aprovado com media YY" considerando o gênero do(a) aluno(a) e se foi aprovado(a) ou reprovado(a)

```python
# Solicita a média para aprovação
media_aprovacao = float(input("Digite a média para aprovação no curso: "))

# Solicita o nome do aluno
nome = input("Digite o nome do aluno: ")

# Solicita o sexo do aluno
sexo = input("Digite o sexo do aluno (M/F): ")

# Solicita as três notas do aluno
nota01 = float(input("Digite a primeira nota do aluno: "))
nota02 = float(input("Digite a segunda nota do aluno: "))
nota03 = float(input("Digite a terceira nota do aluno: "))

# Calcula a média do aluno
media = (nota01 + nota02 + nota03) / 3

# Verifica se o aluno foi aprovado ou reprovado
if media >= media_aprovacao:
    if sexo.upper() == 'M':
        print(f"O aluno {nome} foi aprovado com média {media:.2f}.")
    elif sexo.upper() == 'F':
        print(f"A aluna {nome} foi aprovada com média {media:.2f}.")
else:
    if sexo.upper() == 'M':
        print(f"O aluno {nome} foi reprovado com média {media:.2f}.")
    elif sexo.upper() == 'F':
        print(f"A aluna {nome} foi reprovada com média {media:.2f}.")
```
